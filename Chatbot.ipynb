```python
from langchain_community.llms import Ollama
llm = Ollama(model="llama2")
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a chatbot that specialises in log analysis"),
    ("user", "{input}")
])
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
```

```python
chain = prompt | llm | output_parser
```

```python
from langchain_community.document_loaders.csv_loader import CSVLoader


loader = CSVLoader(file_path='Linux_system_Logs.csv')
data = loader.load()  
```

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = OllamaEmbeddings()
```

```python
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter


text_splitter = RecursiveCharacterTextSplitter()
documents = text_splitter.split_documents(data)
vector = FAISS.from_documents(documents, embeddings)
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings.sentence_transformer import SentenceTransformerEmbeddings
```

```python
embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain

prompt = ChatPromptTemplate.from_template("""Answer the following questions:

<context>  
{context}
</context>

Question: {input}""")

document_chain = create_stuff_documents_chain(llm, prompt)
```

```python
db = Chroma.from_documents(data, embedding_function)
```

```python
from langchain.chains import create_retrieval_chain
retriever = db.as_retriever()
retrieval_chain = create_retrieval_chain(retriever, document_chain)
```

```python
response = retrieval_chain.invoke({"input": "What is the following log file : 2024-04-20 10:15:23 ERROR: CPU overheating detected, temperature exceeds threshold (CPU: 80C)"})
print(response["answer"])
```

```python
response = retrieval_chain.invoke({"input": "Derive insights from the following log file:2024-04-20 10:20:45 ERROR: Memory allocation failed, system running out of memory (Memory: 95%)"})
print(response["answer"])
```
